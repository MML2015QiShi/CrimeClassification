## Description

San Francisco was housing some of the world's most notorious criminals on island of Alcatraz. Given 12 years of crime reports from across all of San Francisco's neighborhood, we must predict the category of crime that occured.

"Dates", "DayOfWeek", "PdDistrict", "Address", "X" and "Y" can be used for modeling.

### Preliminary Analysis

```python
PdDistrict_grouped = pd.crosstab(df_train.PdDistrict,df_train.Category,margins=True)
```

```python
PdDistrict_grouped
```

```python
for key in df_train.Category.unique():
    PdDistrict_grouped[key] = PdDistrict_grouped[key]/PdDistrict_grouped['All']
```

```python
PdDistrict_grouped=PdDistrict_grouped.drop('All',axis=1)
```

```python
PdDistrict_grouped
```

```python
fig = PdDistrict_grouped.plot(kind='bar', stacked = True, figsize=(12, 8))
fig.legend(loc=1,prop={'size':8})
```

Certain department has quite different proportions in the categorical of criminal cases.
